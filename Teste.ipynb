Construção do gráfico x vs t

A função é $$ 
\begin{cases}
x(t) = 6 + 2sen(t) \\
y(t) = -6 - 2cos(t)
\end{cases}
$$

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,9.51,0.01)
x = 6 + 2*np.sin(t)
y = 6 + 2*np.cos(t+np.pi/2)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(t,x,color='r')
ax.plot(t,y,color='b')

plt.show()
```
