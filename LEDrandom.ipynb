```python
from gpiozero import LED, Button
from time import sleep
from signal import pause
import random
```

```python
red = LED(21)
ver = LED(16)

btn1 = Button(20)
btn2 = Button(12)
```

```python
f = list()
n = 1
```

```python
def G1():
    global f, n
    if n > len(f):
        f.append(1)
        n += 1
        print(f)
    else:
        print(f)
    return f, n
```

```python
def G2():
    global f, n
    if n > len(f):
        f.append(2)
        n += 1
        print(f)
    else:
        print(f)
    return f, n
```

```python
def Sinal():
    x = list()
    for i in range(0,6,1):
        y = random.randint(1,2)
        x.append(y)
    return x
        
```

```python
while True:
    t = Sinal()
    f = list()
    red.off()
    ver.off()
    for i in range (0,6,1):
        #print (t[i])
        if t[i] == 1:
            red.on()
            sleep(0.2)
            red.off()
            sleep(0.2)
        else:
            ver.on()
            sleep(0.2)
            ver.off()
            sleep(0.2)
    j = 0        
    while j < 6:
        
        btn1.when_pressed = G1
        btn2.when_pressed = G2
        j = len(f)
        
    if t == f:
        print('Parabéns!')
    else:
        print('Errou!')
    
    sleep(5)   
```
